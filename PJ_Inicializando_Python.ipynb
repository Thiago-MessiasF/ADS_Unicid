```python
#crescimento de capital com juros composto

import matplotlib.pyplot as plt
capital = 1000
taxa = 0.02
meses = 24
valores =[]
for m in range(meses):
 capital = capital * (1+taxa)
 valores.append(capital)
plt.plot(valores)
plt.title("crescimento do ivestimento")
plt.xlabel("meses")
plt.ylabel("capital(R$)")
plt.show()
```

```python
import matplotlib.pyplot as plt
saldo = 0
deposito = 500
meses = 24
historico = []
for m in range(meses):
 saldo += deposito
 historico.append(saldo)
plt.plot(historico)
plt.title("acumulo de poupança")
plt.xlabel("meses")
plt.ylabel("saldo")
plt.show ()
```
